```python
from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform(test.Tweet)

sum_words = words.sum(axis=0)

words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)

frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

frequency.head(30).plot(x='word', y='freq', kind='bar', figsize=(15, 7), color = 'blue')
plt.title("Most Frequently Occuring Words - Top 30")
```

Inference: <br>
Majority words that are seen are emojis which explains the presence of xf0,x9f etc which are twitter encodings for face emoticons and dingbats. <br> 
More about twitter encoding can be found here:https://apps.timwhitlock.info/emoji/tables/unicode

```python
from wordcloud import WordCloud

wordcloud = WordCloud(background_color = 'white', width = 1000, height = 1000).generate_from_frequencies(dict(words_freq))

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.title("WordCloud - Vocabulary from Reviews", fontsize = 22)
```

```python
# collecting the hashtags
import re
def hashtag_extract(x):
    hashtags = []
    
    for i in x:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)

    return hashtags
```

```python
# extracting hashtags from non racist/sexist tweets
HT_regular = hashtag_extract(test['Tweet'])

# extracting hashtags from racist/sexist tweets
HT_negative = hashtag_extract(train['tweet'][train['label'] == 1])

# unnesting list
HT_regular = sum(HT_regular,[])
HT_negative = sum(HT_negative,[])
```

```python
import nltk
a = nltk.FreqDist(HT_regular)
d = pd.DataFrame({'Hashtag': list(a.keys()),
                  'Count': list(a.values())})

# selecting top 15 most frequent hashtags     
d = d.nlargest(columns="Count", n = 15) 
plt.figure(figsize=(16,5))
ax = sns.barplot(data=d, x= "Hashtag", y = "Count")
ax.set(ylabel = 'Count')
plt.title("Top 15 Hashtags around the trending Hashtag")
plt.show()
```

```python
# collecting the hashtags
import re
def tagged_extract(x):
    tagged = []
    
    for i in x:
        ht = re.findall(r"@(\w+)", i)
        tagged.append(ht)

    return tagged
```

```python
# extracting hashtags from non racist/sexist tweets
T_regular = tagged_extract(test['Tweet'])

# extracting hashtags from racist/sexist tweets
T_negative = tagged_extract(train['tweet'][train['label'] == 1])

# unnesting list
T_regular = sum(T_regular,[])
T_negative = sum(T_negative,[])
```

```python
import nltk
a1 = nltk.FreqDist(T_regular)
d1 = pd.DataFrame({'Tags': list(a1.keys()),'Count': list(a1.values())})

# selecting top 20 most frequent tagged accounts     
d1 = d1.nlargest(columns="Count", n = 10) 
plt.figure(figsize=(16,5))
ax1 = sns.barplot(data=d1, x= "Tags", y = "Count")
ax1.set(ylabel = 'Count')
plt.title("Top 20 Tags present in the tweets plotted against their count")
plt.show()
```
